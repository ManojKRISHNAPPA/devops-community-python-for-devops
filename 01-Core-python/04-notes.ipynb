# FILE OPERATION:

r --> read
w --> write 
a --> append
x --> execute



```python
# read file 
with open('devops_community.txt','r') as file:
    content=file.read()
    print(content)
```

```python
## Read line by line
with open('devops_community.txt','r') as file:
    for line in file:
        print(line)
```

```python
## Read line by line
with open('devops_community.txt','r') as file:
    for line in file:
        print(line.strip())
```

```python
with open('devops_community.txt','w') as file:
    file.write("Hello Dheeraj\n")
    file.write ("Good Evng...\n")

```

```python
## write a file without overwriting
with open('devops_community.txt','a') as file:
    file.write("hello Rahul\n")
    file.write("Goodevng...\n")
```

```python
# checking if a path is file or Directory

import os

path="example.txt"
if os.path.isfile(path):
    print(f" this path is {path} is file")
elif os.path.isdir(path):
    print(f"this path is {path} is directory")
else:
    print(f"the path {path} is neither a file or a directory")
```

```python
try:
    result=10/0
except ZeroDivisionError as ex:
    print(ex)
    print(f"please enter number in the denominator which is greater than 0 beacuse it cant be {ex}")
```

```python
try:
    num = int(input("enter your number: "))
    result = 10/num
except ZeroDivisionError as ex:
    print("you cant divide by zero")
else:
    print(f"the result is {result} ")
```

```python
try:
    num = int(input("enter your number: "))
    result = 10/num
except ZeroDivisionError as ex:
    print("you cant divide by zero")
except ValueError as v:
    print(f"you cannot divide with string use {v}")
except Exception as ex:
    print(ex)
else:
    print(f"the result is {result} ")
finally:
    print("you execution is complete")
```
