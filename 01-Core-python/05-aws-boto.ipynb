```python
import boto3

ec2 = boto3.client(
    'ec2',
    region_name='us-east-1',
    aws_access_key_id = "",
    aws_secret_access_key = ""
)
```

```python
response = ec2.describe_instances()
```

```python
print(response)
```

```python
for r in response['Reservations']:
    for i in r['Instances']:
        print(i['InstanceId'], i['State']['Name'])
```

```python
## This is to list the buckets
import boto3

s3 = boto3.client(
    's3',
    region_name='us-east-1',
    aws_access_key_id = "",
    aws_secret_access_key = ""
)

response = s3.list_buckets()

for bucket in response['Buckets']:
    print(bucket['Name'])

```

```python
## Create bucket
import boto3

s3 = boto3.client(
    's3',
    region_name='us-east-1',
    aws_access_key_id = "",
    aws_secret_access_key = ""
)

bucket_name = "devopscommunitypythonfodevops"

s3.create_bucket(Bucket=bucket_name)

print(f" s3 bucket is created: {bucket_name}")
```

```python
file_name = "example.txt"

s3.upload_file(file_name, bucket_name, file_name)

print(f" file is uploaded to {bucket_name}")
```
