```python
# shallow copy
rec={"name":"Manoj","class":10,"grade":7.8}
rec_student=rec
print(rec_student)
print(rec)
```

```python
# shallow copy
rec={"name":"Manoj","class":10,"grade":7.8}
rec_student=rec
print(rec_student)
print(rec)
del rec["name"]
print(rec)
print(rec_student)
```

```python
rec={"name":"Manoj","class":10,"grade":7.8}
rec_student=rec.copy()
print(rec_student)
del rec["name"]
print(rec)
print(rec_student)
```

```python
rec={"name":"Manoj","class":10,"grade":7.8}
rec_student=rec.copy()
print(rec_student)
del rec_student["name"]
print(rec)
print(rec_student)
```
